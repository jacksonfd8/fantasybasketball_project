# Stage of Draft -- Analysis

Context: Auction drafts are not a sprint, like snake drafts. They can take a very long time. I am hoping to answer the following questions: 
- Does stage of draft matter? 
- What is the average cost at certain stages of the draft?
- What strategy can I take to grab some value?

Traditional 12-team league draft, 156 total picks. Divided into 6 "time chunk" draft portions.

```python
### Import data

draft_res = pd.ExcelFile('auctiondraft_results.xlsx')
dr_2018 = pd.read_excel(draft_res, '2018')
dr_2017 = pd.read_excel(draft_res, '2017')
dr_2016 = pd.read_excel(draft_res, '2016')
dr_2015 = pd.read_excel(draft_res, '2015')
```

## Feature Creation -- Methodology

In a league with 12-teams and 13-player roster, there are 156 picks. I decided to segment the data into 6 sizeable chunks.
- Each chunk will contain 26 picks. We will assume each draft stage takes on average 30 minutes.
- In other words, by time the first 26 picks have completed, we would be 30 minutes *into* the draft.
- After 52 picks have completed, we would be 1 hour *into* the draft.

I will create a new column called **draft_stage**.

```python
## Add new column with portion of draft
import numpy as np

dr_2018["draft_stage"] = np.nan
dr_2017["draft_stage"] = np.nan
dr_2016["draft_stage"] = np.nan
dr_2015["draft_stage"] = np.nan

dr_2016
```

```python
## Based on the pick #, assign the "stage" of the draft

def set_stage(df):

    for i in range(len(df)): 
        if df["Pick"][i] <=26: 
            df["draft_stage"][i] ="1st"

        elif df["Pick"][i] <= 52: 
            df["draft_stage"][i] ="2nd"

        elif df["Pick"][i] <= 78: 
            df["draft_stage"][i] ="3rd"

        elif df["Pick"][i] <= 104: 
            df["draft_stage"][i] ="4th"

        elif df["Pick"][i] <= 130: 
            df["draft_stage"][i] = "5th"

        else: 
            df["draft_stage"][i] = "6th"

set_stage(dr_2018)
set_stage(dr_2017)
set_stage(dr_2016)
set_stage(dr_2015)
```

## Add rankings, Merge dataframes from same year

Data sources:
- Draft Results from Yahoo! spanning from 2015-2018 
- Basketball Monster Player Rankings from 2015-2018 

Currently, the Draft Results data has the player's position in the **Player** field, we'll need to remove the paranthesis from this field.

```python
dr_2018
```

```python
## Data cleaning

## Creates new 'Name' col that eliminates the 
## paranetheses in 'Player' col

def elim_par(df):
    df['Name'] = np.nan
    name_list = []

    for row in df['Player']:
        new_str = row.partition("(")[0]
        new_str = new_str.rstrip()
        name_list.append(new_str)

    df['Name'] = name_list
```

```python
elim_par(dr_2018)
elim_par(dr_2017)
elim_par(dr_2016)
elim_par(dr_2015)
```

```python
def round_average(df):
    
    count_1st = 0
    total_1st = 0 
    count_2nd = 0
    total_2nd = 0 
    count_3rd = 0
    total_3rd = 0 
    count_4th = 0
    total_4th = 0 
    count_5th = 0
    total_5th = 0 
    count_6th = 0
    total_6th = 0 

    for i in range(len(df)): 
        if df["draft_stage"][i] == '1st': 
            total_1st += df["Cost"][i] 
            count_1st += 1      

        elif df["draft_stage"][i] == '2nd': 
            total_2nd += df["Cost"][i] 
            count_2nd += 1 


        elif df["draft_stage"][i] == '3rd':  
            total_3rd += df["Cost"][i] 
            count_3rd += 1 


        elif df["draft_stage"][i] == '4th': 
            total_4th += df["Cost"][i] 
            count_4th += 1 


        elif df["draft_stage"][i] == '5th': 
            total_5th += df["Cost"][i] 
            count_5th += 1 

        else: 
            total_6th += df["Cost"][i] 
            count_6th += 1 

    print("Avg cost of player in 1st stage:", round(total_1st/count_1st))
    print("Avg cost of player in 2nd stage:", round(total_2nd/count_2nd))
    print("Avg cost of player in 3rd stage:", round(total_3rd/count_3rd))
    print("Avg cost of player in 4th stage:", round(total_4th/count_4th))
    print("Avg cost of player in 5th stage:", round(total_5th/count_5th))
    print("Avg cost of player in 6th stage:", round(total_6th/count_6th))
```

```python
print("2018")
round_average(dr_2018)
print("\n2017")
round_average(dr_2017)
print("\n2016")
round_average(dr_2016)
print("\n2015")
round_average(dr_2015)
```

## Takeaways
- People tend to nominate more expensive players in the first stage (first 30 minutes) of a draft. In fact, people are more trigger-happy and loose with their budget in the beginning. A solid strategy would be to nominate overhyped, lower-ranked players in the earlier stages of the draft so you can get other managers to waste money, giving you more buying power as the draft progresses.
- The 3rd-5th stages offer a ton of valuable players for a small percentage of your budget. Stay alert during this timeframe!
 - ($12.5 of 200 = 6% of total budget, 7/200 = 3.5%, 4.75/200 = 2%)
